```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("climate_change_dataset.csv")

# ---------------------------
# Task 2: Dataset Investigation
# ---------------------------

print("Number of Records:", df.shape[0])
print("Number of Variables:", df.shape[1])

print("\nVariable Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

# ---------------------------
# Task 4: Data Cleaning
# ---------------------------

print("\nMissing Values:")
print(df.isnull().sum())

print("\nTotal Missing Values:")
print(df.isnull().sum().sum())

print("\nDuplicate Records:")
print(df.duplicated().sum())

# Detect Outliers
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"\n{col}")
    print("Outliers:", len(outliers))

# ---------------------------
# Task 5: Descriptive Statistics
# ---------------------------

avg_temp        = np.array(df["Avg Temperature (°C)"])
co2             = np.array(df["CO2 Emissions (Tons/Capita)"])
sea_level       = np.array(df["Sea Level Rise (mm)"])
rainfall        = np.array(df["Rainfall (mm)"])
population      = np.array(df["Population"])
renewable       = np.array(df["Renewable Energy (%)"])
extreme_weather = np.array(df["Extreme Weather Events"])
forest_area     = np.array(df["Forest Area (%)"])

columns = {
    "Avg Temperature (°C)"        : avg_temp,
    "CO2 Emissions (Tons/Capita)" : co2,
    "Sea Level Rise (mm)"         : sea_level,
    "Rainfall (mm)"               : rainfall,
    "Population"                  : population,
    "Renewable Energy (%)"        : renewable,
    "Extreme Weather Events"      : extreme_weather,
    "Forest Area (%)"             : forest_area,
}

for name, arr in columns.items():
    print(f"\n========== {name} ==========")
    print("Mean               :", arr.mean())
    print("Median             :", np.median(arr))
    print("Mode               :", float(pd.Series(arr).mode().iloc[0]))
    print("Min                :", np.min(arr))
    print("Max                :", np.max(arr))
    print("Range              :", np.max(arr) - np.min(arr))
    print("Variance           :", np.var(arr, ddof=1))
    print("Standard Deviation :", np.std(arr, ddof=1))
    print("Q1 (25th pctile)   :", np.percentile(arr, 25))
    print("Q2 (50th pctile)   :", np.percentile(arr, 50))
    print("Q3 (75th pctile)   :", np.percentile(arr, 75))
    print("P10                :", np.percentile(arr, 10))
    print("P90                :", np.percentile(arr, 90))
```
